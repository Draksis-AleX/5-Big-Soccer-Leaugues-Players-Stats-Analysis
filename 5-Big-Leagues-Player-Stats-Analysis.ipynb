```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset

# https://www.kaggle.com/datasets/mamounkabbaj/2023-2024-big-5-european-soccer-player-statistics?resource=download
stats_df = pd.read_csv("big_5_players_stats_2023_2024.csv")
stats_df.head()
```

```python
stats_df.info()
```

```python
stats_df.describe()
```

```python
stats_df.isna().sum()
```

```python
# Create average minutes played per match column

stats_df['Per Match Min'] = (stats_df['Playing Time_Min'] / stats_df['Playing Time_MP']).astype(int)

# Create average goals per match column

stats_df['Per Match Gls'] = (stats_df['Performance_Gls'] / stats_df['Playing Time_MP']).astype(float)

stats_df.head()
```

---

```python
# Total Goals

total_goals = stats_df['Performance_Gls'].sum()
print(total_goals)
```

```python
# Total Goal per League

stats_df_per_league = stats_df.groupby('Competition')
total_goals_per_league = stats_df_per_league['Performance_Gls'].sum()
print(total_goals_per_league)
```

---

## Penalties

```python
# Total Penalty Goals

total_pen_goals = stats_df['Performance_PK'].sum()
print(total_pen_goals)
```

```python
# Total Penalty Goals per League

total_pen_goals_per_league = stats_df_per_league['Performance_PK'].sum()
print(total_pen_goals_per_league)
```

```python
# Total Penalties Attempts

total_pen_attempts = stats_df['Performance_PKatt'].sum()
print(total_pen_attempts)
```

```python
# Total Penalty Attempts per League

total_pen_attempts_per_league = stats_df_per_league['Performance_PKatt'].sum()
print(total_pen_attempts_per_league)
```

```python
# Scored vs Missed Penalties pie chart

plt.figure(figsize=(13, 6))
pk_not_scored = stats_df['Performance_PKatt'].sum() - stats_df['Performance_PK'].sum()
data = [pk_not_scored, total_pen_goals]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show
```

```python
# Scored vs Missed Penalties pie chart per league

# Group by leagues
stats_df_per_league = stats_df.groupby('Competition')

total_pen_goals_per_league = stats_df_per_league['Performance_PK'].sum()
total_pen_attempts_per_league = stats_df_per_league['Performance_PKatt'].sum()

num_leagues = len(total_pen_goals_per_league)

fig, axes = plt.subplots(1, num_leagues, figsize=(30, 12), squeeze=False)

# Create a pie chart for each league
for i, league in enumerate(total_pen_goals_per_league.index):
    pk_not_scored = total_pen_attempts_per_league[league] - total_pen_goals_per_league[league]
    total_pen_goals = total_pen_goals_per_league[league]
    
    data = [pk_not_scored, total_pen_goals]
    labels = ['Penalties missed', 'Penalties Scored']
    color = sns.color_palette('Set2')
    
    axes[0, i].pie(data, labels=labels, colors=color, autopct='%.0f%%')
    axes[0, i].set_title(f'Penalties Distribution in {league}')

plt.tight_layout()
plt.show()
```

```python
# Clubs with least players in their squad

stats_df['Squad'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette("viridis"))
```

```python
# Players based on age group

Under20 = stats_df[stats_df['Age'] <= 20]
Age20_25 = stats_df[(stats_df['Age'] > 20) & (stats_df['Age'] <= 25)]
Age25_30 = stats_df[(stats_df['Age'] > 25) & (stats_df['Age'] <= 30)]
Over30 = stats_df[stats_df['Age'] > 30]
```

```python
x = np.array([Under20['Player'].count(), Age20_25['Player'].count(), Age25_30['Player'].count(), Over30['Player'].count()])
mylabels = ["<= 20", "20 - 25", "25 - 30", " > 30"]
plt.title("Total Players by Age Group", fontsize = 20)
plt.pie(x, labels=mylabels, autopct="%.1f%%")
plt.show()
```

```python
# Total under 20 players in each club

players_under_20 = stats_df[stats_df['Age'] < 20]
players_under_20['Squad'].value_counts().nlargest(10).plot(kind='bar', color=sns.color_palette("cubehelix"))
```
